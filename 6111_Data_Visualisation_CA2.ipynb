# **Data Visualization**

# **Finding the Trends & Patterns among the Attributes**

```python
correlation_matrix = df.corr()
plt.figure(figsize=(11, 6))

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.title('Correlation Heatmap')

plt.xticks(fontsize=9,rotation=30)
plt.yticks(fontsize=9)

plt.show()
```

* **With this we can conclude that there is no such strong correlation among the attributes.**

# **Finding the patterns between Pixels & Storage of Cameras**

```python
plt.title("Relation of Pixel & Storage")
sns.barplot(x=df['Effective_pixels'],y=df['Storage_included(TB)'])
plt.show()
```

* **We can conclude that if we increase the storage & pixel we get the better picture.**

## **Finding the release year of most amount of camera.**

```python
plt.figure(figsize=(9,8))
df['Release_date'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True)
plt.title('Manufacturing of Cameras')
plt.show()
```

* **With this we can conclude that with the increase in years the cameras where manufactured in more numbers.**
* **This also concludes that the cameras where more used with the development of the cameras.**

# **Finding the patterns between the Normal Focus and Macro Focus Range**

```python
plt.figure(figsize=(8,6))
plt.title('Pattern of Normal and Macro Focus Range')
sns.lineplot(x="Macro_focus_range",y="Normal_focus_range",data=df,color='cyan')
plt.xlabel('Macro_focus_range')
plt.ylabel('Normal_focus_range')
plt.show()
```

* **With this we can conclude that the Macro Focus Range increases after the certain limit of Normal Focus Range.**
* **Hence to take the picture of macro object we need certain limit of Normal Focus Range.**

# **Finding the Max. resolutions of Cameras made of.**

```python
plt.hist(df['Max_resolution(PPI)'],bins=12)
plt.xlabel("Max. Resolution(Pixel Per Inches)")
plt.ylabel("No. of cameras")
plt.title("Maximum Resolution of Cameras made")
plt.show()
```

* **With this we can conclude that most cameras where made up of approx. to 3000 PPI for more clarity in the picture.**

# **Finding the price ranges of Cameras with respect to the Dimensions.**

```python
plt.figure(figsize=(8,6))
plt.title('Price of Cameras with respect to Dimension')
sns.scatterplot(data=df,x='Dimensions',y='Price',color='red')
plt.show()
```

* **With this we can conclude that the most camera where under range 0-2000.**
* **There are some cameras which price is greater than 2000 & some are of price Rs. 8000 also.**

# **Finding the trends of Storage of Cameras & Weight of Batteries**

```python
plt.figure(figsize=(5,7))
sns.jointplot(data=df,x='Storage_included(TB)',y='Weight(inc. batteries)',color='orange')
plt.show()
```

* **With this we can conclude that the most camera of storage under 100 TB where having battery's weight of 750g.**

# **Finding the No. of pixels each camera where made of.**

```python
plt.figure(figsize=(5,5))
plt.title("No. of Cameras made with respect to pixels")
sns.countplot(x='Effective_pixels',data=df)
plt.show()
```

* **With this we can conclude that the most cameras where made up of 3 MP which counts to 200 otherwise camera where made up of range 4-8 MP.**

# **Relationship between Max. Resolution and Low Resolution of Camera**

```python
plt.figure(figsize=(9,6))
sns.jointplot(data=df,x='Max_resolution(PPI)',y='Low_resolution(PPI)',kind='hex',color='purple')
plt.show()
```

* **With this we can conclude that the Min. resolution of camera is approx. 750 PPI and Max. resolution of camera is approx. 4000 PPI.**
