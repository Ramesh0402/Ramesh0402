##**Introduction**

The Manufacturing Defect Dataset is a comprehensive collection of data pertaining to various aspects of a manufacturing process, aimed at identifying and analyzing defects within the production line. The dataset contains a diverse set of features that capture different dimensions of the manufacturing environment, including production volume, costs, quality metrics, maintenance activities, and more. This dataset is particularly valuable for understanding the factors that contribute to manufacturing defects and for developing predictive models to improve quality control.

# **Objectives**

The primary objective of this dataset is to facilitate the analysis of manufacturing defects by examining the relationships between various production factors and the occurrence of defects. By exploring these relationships, we aim to:
1.  
Identify key factors that contribute to defects.
2.  
Develop predictive models to forecast defects based on production data.
3.
Implement data-driven strategies to reduce defect rates and improve overall production quality.

# **Dataset Features**


The dataset comprises the following features:
1.  **ProductionVolume :** The number of units produced during a specific time period.

1.  **ProductionCost :** The cost incurred in producing the units.

1.   **SupplierQuality :** A score indicating the quality of materials provided by suppliers.

1.   **DeliveryDelay :** The number of delays in the delivery of materials or products.

1.   **DefectRate :** The percentage of defective units produced.

1.  **QualityScore :** An overall score representing the quality of the production process.

1.   **MaintenanceHours :** The number of hours spent on maintenance activities.

1.   **DowntimePercentage :** The percentage of time the production line was down.

1.   **InventoryTurnover :** The rate at which inventory is used and replenished.

1.   **StockoutRate :** The rate at which inventory stockouts occur.

1.   **WorkerProductivity :** A score indicating the productivity of the workers.

1.   **SafetyIncidents :** The number of safety incidents reported.

1.   **EnergyConsumption :** The amount of energy consumed during production.

1.   **EnergyEfficiency :** A measure of how efficiently energy is used in the production process.

1.   **AdditiveProcessTime :** The time taken for additive manufacturing processes.

1.   **AdditiveMaterialCost :** The cost of materials used in additive manufacturing.

1.   **DefectStatus :** A binary indicator of whether a defect was present (1) or not (0).

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.inspection import permutation_importance

# Feature Importance using RandomForestClassifier
# Splitting the data
X = df.drop('DefectStatus', axis=1)
y = df['DefectStatus']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardizing the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Training the RandomForest model
model = RandomForestClassifier(random_state=42)
model.fit(X_train_scaled, y_train)

# Getting feature importances
importances = model.feature_importances_
feature_names = X.columns
forest_importances = pd.Series(importances, index=feature_names)

# Plotting feature importances
plt.figure(figsize=(12, 8))
forest_importances.sort_values().plot(kind='barh')
plt.title('Feature Importances from RandomForestClassifier')
plt.xlabel('Mean Decrease in Impurity')
plt.ylabel('Feature')
plt.show()

# Permutation Importance
perm_importance = permutation_importance(model, X_test_scaled, y_test, n_repeats=10, random_state=42)
perm_sorted_idx = perm_importance.importances_mean.argsort()

plt.figure(figsize=(12, 8))
plt.boxplot(perm_importance.importances[perm_sorted_idx].T, vert=False, labels=X.columns[perm_sorted_idx])
plt.title('Permutation Importances (test set)')
plt.xlabel('Mean Importance')
plt.show()
```

* This shows the relationship between the defect status and the other factors affecting the defective rate of the production.

# **Model Building**

# **Logistic Regression**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Load the dataset
data = pd.read_csv('/content/manufacturing_defect_dataset.csv')

# Feature and target variable
X = data.drop('DefectStatus', axis=1)
y = data['DefectStatus']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Logistic Regression model
model = LogisticRegression()
model.fit(X_train_scaled, y_train)

# Predictions
y_pred = model.predict(X_test_scaled)

# Evaluation
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)

print(f'Accuracy: {accuracy:.2f}')
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1 Score: {f1:.2f}')
print(f'ROC AUC Score: {roc_auc:.2f}')
```

* In this model building since the last column is of Defective Status which consist of boolean value.
* Hence, based on that I had decided to go through the Logistic Regression Algorithm.
* Based on that we could the **Accuracy 0.88** and **Precision 0.90**.
* This shows that our model is showing almost good result with less chance of error.
* Hence, this model will predict the defective product in the manufacturing process.
