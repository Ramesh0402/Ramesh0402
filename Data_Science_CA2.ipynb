# **Data Preprocessing**

* **Handling Null Values**

```python
# Finding the count of null values of each dimension
df.isnull().sum()
```

```python
# Finding the values of "processor_gnrtn" column
df["processor_gnrtn"].value_counts()
```

 * Since the processor generation are varying according to users satisfaction. So, we can fill the null values with the term "Missing".

```python
# Fill df["processor_gnrtn"] null values with "Missing"

df["processor_gnrtn"].fillna("Missing", inplace=True)
```

```python
# Checking the null values whether present in the processor_gnrtn column
df["processor_gnrtn"].isnull().sum()
```

```python
# Finding the values of "processor_name" column
df["processor_name"].value_counts()
```

```python
# To show the df["processor_name"].isnull record
df[df["processor_name"].isnull()]
```

* Since, all the values in that record have null values. We can decide of removing that record.

```python
# Dropping the irrelevant record of "processor_name" column.
df = df.drop(df[df["processor_name"].isnull()].index)
```

```python
#Checking the null values in the "processor_name" column
df["processor_name"].isnull().sum()
```

```python
df["os"].value_counts()
```

```python
# Fill df["os"] null values with "no os"
df["os"].fillna("no os", inplace=True)
```

```python
# Checking the null values in the "os" column
df["os"].isnull().sum()
```

```python
# Finding the values of "ram_gb" column
df["ram_gb"].value_counts()
```

 * After analysing, we can replace the null values with the term "Missing".

```python
# Fill null values of df["ram_gb"] with "Missing"
df["ram_gb"].fillna("Missing", inplace=True)
```

```python
#Since the RAM GB is always in the numerical format. So, we can replace the irrelevant data to "Missing".

df["ram_gb"] = df["ram_gb"].replace("15.6", "16")
df["ram_gb"] = df["ram_gb"].replace("5", "4")
df["ram_gb"] = df["ram_gb"].replace("NVIDIA", "Missing")
df["ram_gb"] = df["ram_gb"].replace("Dual", "Missing")
df["ram_gb"] = df["ram_gb"].replace("Acer", "Missing")
df["ram_gb"] = df["ram_gb"].replace("Pre-installed", "Missing")
df["ram_gb"] = df["ram_gb"].replace("Full", "Missing")
df["ram_gb"] = df["ram_gb"].replace("Intel", "Missing")
df["ram_gb"] = df["ram_gb"].replace("Access", "Missing")
df["ram_gb"].value_counts()
```

```python
# Checking the null values in the "ram_gb" column
df["ram_gb"].isnull().sum()
```

```python
# Finding the values of "ram_type" column
df["ram_type"].value_counts()
```

```python
# Fill df["ram_type"] null values with "Missing"
df["ram_type"].fillna("Missing", inplace=True)
```

```python
# Replace the irrelevant data to relevant data
df["ram_type"] = df["ram_type"].replace("Cooling","Missing")
df["ram_type"] = df["ram_type"].replace("Master","Missing")
df["ram_type"] = df["ram_type"].replace("Windows","Missing")
df["ram_type"] = df["ram_type"].replace("LED","Missing")
df["ram_type"] = df["ram_type"].replace("i5","Missing")
df["ram_type"] = df["ram_type"].replace("Apps","Missing")
df["ram_type"].value_counts()
```

```python
# Finding the "processor_name" values to find the irrelevant data
df["processor_name"].value_counts()
```

```python
# Replacing the irrelevant data to relevant data
df["processor_name"] = df["processor_name"].replace("i7","Core i7")
df["processor_name"] = df["processor_name"].replace("i5","Core i5")
df["processor_name"] = df["processor_name"].replace("5","Core i5")
df["processor_name"] = df["processor_name"].replace("Core m3","Core i3")
df["processor_name"] = df["processor_name"].replace("M1 Max","M1 Pro")
df["processor_name"] = df["processor_name"].replace("Core","Core i3")
df["processor_name"].value_counts()
```

```python
# To find the no. of model of each brand.
df.groupby("brand")["model"].nunique()
```

```python
# To show the name of the brand model
df.groupby("brand")["model"].unique().agg(list)
```

```python
#Replacing the irrelevant name to its proper name.
df["model"] = df["model"].replace("Vivobook","VivoBook")
df["model"] = df["model"].replace("Ideapad","IdeaPad")
df["model"] = df["model"].replace("2020","MacBook")
df["model"] = df["model"].replace("2021","MacBook")
df.groupby("brand")["model"].unique().agg(list)
```

```python
#To find the duplicate values
df.duplicated().sum()
```

```python
# Dropping the duplicated values
df=df.drop_duplicates()
```

```python
#At last finding the null values in the dataset
df.isnull().sum().sum()
```

```python
k= 11
cols = correlation.nlargest(k,'latest_price')['latest_price'].index
print(cols)
cm = np.corrcoef(df[cols].values.T)
mask = np.triu(np.ones_like(df.corr()))
f , ax = plt.subplots(figsize = (12,10))
sns.heatmap(cm,mask=mask, vmax=.8, linewidths=0.01,square=True,annot=True,cmap='inferno',
            linecolor="white",xticklabels = cols.values ,annot_kws = {'size':12},yticklabels = cols.values)
```

# **Conclusion**

For many, the cost of a laptop plays a decisive role, while important parameters are overlooked. As a result â€” regrets after the first days of operation. At the same time, even a famous brand and a high price do not guarantee that you will get what really suits you. Often, even with a huge choice, users are forced to compromise, so long as a couple or three important parameters match.

Therefore, first of all, it is worth deciding for what purpose you are purchasing a laptop. This will depend on the size and resolution of the screen, the processor, video card and the amount of RAM.

* The most common problem statement is a machine for working with documents, multimedia and surfing the Internet. For these purposes, a CPU with average performance and an integrated video card is suitable. When choosing a model, many users are guided by the low price of the device, losing sight of the fact that the computer has an outdated processor with low efficiency. At first, the weak heart of the computer will demonstrate acceptable speed in office applications and when surfing the net. However, as soon as Windows runs an update in the background, a virus scan, or simply copying files to an external drive, the machine will turn into a prehistoric turtle. The Intel Core i5 and i7 CPU models will allow you to have some margin in computing power - they are used in powerful workstations and gaming mobile PCs. Also, on sale you can find laptops with Ryzen 5 processors from AMD, which have good integrated graphics on board, which is approximately on par with the GT 1030.

* Recently, manufacturers have been actively exploiting the trend of screens with 4K resolution (3840 Ã— 2160). In our opinion, it is a mistake to succumb to marketing traps and pay attention to such offers in portable devices. The fact is that the human eye already has difficulty distinguishing pixels on a 23-inch monitor with Full HD resolution from a distance of 40 cm. At 17 or 15 inches, you definitely wonâ€™t feel an increase in image quality. At the same time, a 4K monitor will show too small elements of the Windows interface and programs, which will only cause discomfort. It should be borne in mind that a more detailed display of graphics in 4K takes resources from the video card and CPU. Therefore, the best screen resolution for a laptop is 1920x1080. If you want more, connect your computer to a large UHD monitor and enjoy an ultra-detailed picture on it.

* Many manufacturers, in an effort to win in the competition, save on additional slots in laptops by installing only one storage device. Choosing such a model, for example, with a fast SSD, you risk being left without a system and without personal data.

* The option with one HDD is also not very preferable, because. This type of media today demonstrates very low data exchange rates. Your system will be slow to boot, and applications will be annoying with the slowness of launching and processing data. Therefore, when choosing a device, give preference to the SSD + HDD dual disk system.

* For office work and surfing the Web, it is desirable to have at least 4 GB of RAM. For graphics and games, 8 GB is enough. Complex tasks may require 16 or even 32 GB of RAM. In the latter case, it is important to have both a very productive and expensive CPU. Before buying, ask how the memory is installed - one module or two equal parts, i.e. in dual-channel mode, which somewhat speeds up data processing.
